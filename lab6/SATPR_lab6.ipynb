Відстані між базами і споживачами

```python
distances = [
    [68, 72, 75, 83],
    [56, 60, 58, 63],
    [38, 40, 35, 45],
    [47, 42, 40, 45]
]
```

```python
num_bases = len(distances)
num_consumers = len(distances[0])
```

Створення задачі лінійного програмування

```python
model = LpProblem("Minimize_Transportation_Cost", LpMinimize)
```

Створення змінних вирішення

```python
x = LpVariable.dicts("Route", ((i, j) for i in range(num_bases) for j in range(num_consumers)), 0, 1, LpMinimize)
```

Визначення функції витрат

```python
model += lpSum(distances[i][j] * x[i, j] for i in range(num_bases) for j in range(num_consumers)), "Total_Transportation_Cost"
```

Умова, що кожне замовлення повинно бути виконане однією з баз

```python
for j in range(num_consumers):
    model += lpSum(x[i, j] for i in range(num_bases)) == 1, f"Consumer_{j + 1}_Assignment"
```

Умова, що кожна база може виконувати лише одне замовлення

```python
for i in range(num_bases):
    model += lpSum(x[i, j] for j in range(num_consumers)) == 1, f"Base_{i + 1}_Assignment"
```

Розв'язання задачі

```python
model.solve()
```

Виведення результатів

```python
print(f"Optimal Transportation Cost: {model.objective.value()}")
print("Optimal Assignment:")
for var in model.variables():
    if var.varValue == 1:
        print(f"{var.name} -> {var.value()}")
```
